**Checkig missing values percentage**

```python
missing = pd.DataFrame({
    "Missing_Count": df.isnull().sum(),
    "Missing_%": (df.isnull().sum() / len(df) * 100).round(2)
}).sort_values(by="Missing_%", ascending=False)

missing
```

**Droping Xylene**

```python
df.drop(columns=["Xylene"], inplace=True)
print("New shape:", df.shape)
df.columns
```

**Handling missing values**

```python
pollutant_cols = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene']

# city-wise median fill
df[pollutant_cols] = df.groupby("City")[pollutant_cols].transform(
    lambda x: x.fillna(x.median())
)

# check missing again
df[pollutant_cols].isnull().sum()
```

**Global Median Backup**

```python
# fill remaining nulls with overall median (all cities)
for col in pollutant_cols:
    df[col] = df[col].fillna(df[col].median())

# check again
df[pollutant_cols].isnull().sum()
```

**Handle Missing AQI + AQI_Bucket**

```python
print("Before:", df.shape)

df = df.dropna(subset=["AQI", "AQI_Bucket"])

print("After:", df.shape)

# confirm no missing
df[["AQI", "AQI_Bucket"]].isnull().sum()
```

**Outlier Handling**

```python
numeric_cols = pollutant_cols + ["AQI"]

# cap at 1st and 99th percentile
for col in numeric_cols:
    low = df[col].quantile(0.01)
    high = df[col].quantile(0.99)
    df[col] = df[col].clip(low, high)

df[numeric_cols].describe()
```

## Season column

```python
def get_season(month):
    if month in [12,1,2]:
        return "Winter"
    elif month in [3,4,5]:
        return "Summer"
    elif month in [6,7,8,9]:
        return "Monsoon"
    else:
        return "Post-Monsoon"

df["Season"] = df["Month"].apply(get_season)

df["Season"].value_counts()
```

### Top Pollutant per City

```python
pollutant_cols = ["PM2.5","PM10","NO2","SO2","CO","O3"]

city_pollutants = df.groupby("City")[pollutant_cols].mean()

# find top pollutant for each city
top_pollutant = city_pollutants.idxmax(axis=1)

top_pollutant.head(10)
```

### Almost every city → PM10

This does NOT mean PM10 is actually the dominant pollutant in reality.

```python
df.groupby("City")[pollutant_cols].mean().idxmax(axis=1)
```

## Correct Approach

```python
top_pollutant_corr = {}

for city in df["City"].unique():
    city_df = df[df["City"] == city]
    
    corr = city_df[["PM2.5","PM10","NO2","SO2","CO","O3","AQI"]].corr()["AQI"].drop("AQI")
    
    top_pollutant_corr[city] = corr.idxmax()

top_pollutant_corr = pd.Series(top_pollutant_corr)

top_pollutant_corr.head(10)
```

## What this actually means

### PM2.5 is the most dominant pollutant (overall)  
Even though not every city shows it in top 10, across all cities:  
PM2.5 will dominate most cities  
This matches the earlier result:  
PM2.5 correlation with AQI = 0.76 (highest)  

## Final Conclusion

The analysis reveals that PM2.5 is the most significant pollutant affecting AQI across most cities, followed by PM10 and gaseous pollutants such as NO₂ and CO. This indicates that particulate matter and combustion-related emissions are the primary drivers of urban air pollution in India

## ML PHASE

Defining X and Y

## Step 1: Logistic Regression

```python
# features (inputs)
X = df[["PM2.5","PM10","NO","NO2","NOx","NH3","CO","SO2","O3","Benzene","Toluene"]]

# target (output)
y = df["AQI_Bucket"]

print(X.shape, y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)

# check mapping
dict(zip(le.classes_, le.transform(le.classes_)))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)

print(X_train.shape, X_test.shape)
```

## Step 2

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# create model
model = LogisticRegression(max_iter=1000)

# train
model.fit(X_train, y_train)

# predict
y_pred = model.predict(X_test)

# accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# detailed report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

## Step 2.1

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Re-train Logistic Regression

```python
model = LogisticRegression(max_iter=2000)

model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

### Key insights

Feature scaling significantly improved model performance by stabilizing optimization and improving classification across minority classes.

## STEP 3: Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_rf))
```

### Key insights

Random Forest outperformed Logistic Regression by capturing non-linear relationships in air pollution data and providing better classification accuracy and balanced performance across AQI categories.

## STEP 4: Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()

gb.fit(X_train, y_train)

y_pred_gb = gb.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_gb))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_gb))
```

# PLANT SYSTEM

## STEP 1: Create Foundation Mapping

```python
plant_mapping = {
    "PM2.5": ["Bamboo", "Neem", "Peepal"],
    "PM10": ["Bamboo", "Ashoka"],
    "CO": ["Snake Plant", "Money Plant"],
    "NO2": ["Neem", "Ashoka"],
    "SO2": ["Neem", "Cassia"],
    "O3": ["Banyan"]
}
```

## Summary

Different plant species have varying abilities to absorb pollutants based on their leaf structure, surface area, and gas exchange properties. ## City → Pollutant → Plant Recommendation System

```python
def recommend_for_city(city):
    
    # get top pollutant for the city
    pollutant = top_pollutant_corr.get(city)
    
    # get plants for that pollutant
    plants = plant_mapping.get(pollutant, "No data available")
    
    return {
        "City": city,
        "Dominant Pollutant": pollutant,
        "Recommended Plants": plants
    }
```

## STEP 4: AQI reduction simulation 

### If we plant trees → pollution decreases → AQI improves
Increasing green cover reduces pollution by a small % locally

Assumption:
10% increase in green cover → 3% pollution reduction  

The relationship between green cover and pollution reduction is based on findings from environmental studies, which indicate that urban vegetation can reduce particulate matter and gaseous pollutants. Due to variability in environmental conditions, an exact relationship cannot be established. Therefore, a conservative proportional model was adopted, assuming that a 10% increase in green cover results in an approximate 3% reduction in pollution levels. This approach provides a realistic estimation for demonstrating potential environmental impact.  

You can mention:  

“Based on environmental studies (approximation)”

```python
def simulate_reduction(city, green_increase_percent=10):
    
    pollutant = top_pollutant_corr.get(city)
    
    current_aqi = df[df["City"] == city]["AQI"].mean()
    
    # calculate pollution reduction
    pollution_reduction = green_increase_percent * 0.3
    
    # new AQI
    new_aqi = current_aqi * (1 - pollution_reduction/100)
    
    return {
        "City": city,
        "Dominant Pollutant": pollutant,
        "Current AQI": round(current_aqi, 2),
        "Estimated AQI after plantation": round(new_aqi, 2),
        "Green Cover Increase (%)": green_increase_percent,
        "Estimated Pollution Reduction (%)": pollution_reduction
    }
```

```python
simulate_reduction("Delhi")
```

The reduction values are estimated based on literature and simplified assumptions, and are intended to demonstrate potential impact rather than exact real-world outcomes.

## FINAL STEP: Wrap Everything into a Clean System

```python
def final_solution(city, green_increase_percent=10):
    
    # dominant pollutant
    pollutant = top_pollutant_corr.get(city)
    
    # plant recommendation
    plants = plant_mapping.get(pollutant, "No data available")
    
    # current AQI
    current_aqi = df[df["City"] == city]["AQI"].mean()
    
    # pollution reduction
    pollution_reduction = green_increase_percent * 0.3
    
    # new AQI
    new_aqi = current_aqi * (1 - pollution_reduction/100)
    
    return {
        "City": city,
        "Dominant Pollutant": pollutant,
        "Recommended Plants": plants,
        "Current AQI": round(current_aqi, 2),
        "Estimated AQI after plantation": round(new_aqi, 2),
        "Green Cover Increase (%)": green_increase_percent,
        "Estimated Pollution Reduction (%)": pollution_reduction
    }
```

```python
final_solution("Delhi", 20)
```
